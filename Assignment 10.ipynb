```python
'''Q1. What is multithreading in python? Why is it used? Name the module used to handle threads in python?'''
#Ans..

'''Multithreading in Python refers to the ability of a program to execute multiple threads concurrently.

Multithreading is used in scenarios where there are multiple tasks or operations that can be executed
independently and concurrently. It is particularly useful in situations where there are blocking or
time-consuming operations, such as network requests or I/O operations. 

The threading module is used to handle threads in Python.'''
```

```python
'''Q2. Why threading module used? Write the use of the following functions?
1. activeCount()
2. currentThread()
3. enumerate()'''

#Ans..

'''The threading module is used in Python to handle threads and provide a high-level interface for creating
and managing them. It offers various functions and methods to control thread execution and synchronization.

1. activeCount(): This function returns the number of Thread objects currently alive. It is useful to 
check the number of active threads in a program.

2. currentThread(): This function returns the current Thread object corresponding to the caller's
thread of control. It can be used to get information about the current thread, such as its name or 
identification number.

3. enumerate(): This function returns a list of all active Thread objects currently alive. It is useful
to get a list of all running threads in a program. Each Thread object in the list can be accessed to
retrieve information about the thread, such as its name or identification number.'''
```

```python
'''Q3. Explain the following functions:
1. run()
2. start()
3. join()
4. isAlive()'''

#Ans..

'''1. run(): This method is the entry point for the thread's activity. It is called when the start() method
is invoked on a Thread object.

2. start(): This method starts the execution of the thread. It creates a new thread of control and calls 
the run() method of the thread. Once the start() method is called, the thread enters the "started" state
and begins executing concurrently with other threads.

3. join(): This method blocks the calling thread until the thread whose join() method is called completes
its execution. It allows one thread to wait for another thread to finish before proceeding further

4. isAlive(): This method returns True if the thread is currently alive (i.e., in the "started" state)
and False otherwise. A thread is considered alive from the moment it is started using the start() method
until it completes its execution or is terminated. This method can be used to check if a thread is still
running before performing any actions based on its completion status.'''
```

```python
'''Q4. Write a python program to create two threads. Thread one must print the list of squares and thread
two must print the list of cubes.'''

#Ans..

import threading

def print_squares():
    squares = [x**2 for x in range(1, 6)]
    for num in squares:
        print(num)

def print_cubes():
    cubes = [x**3 for x in range(1, 6)]
    for num in cubes:
        print(num)

if __name__ == "__main__":
    thread1 = threading.Thread(target=print_squares)
    thread2 = threading.Thread(target=print_cubes)

    thread1.start()
    thread2.start()

    thread1.join()
    thread2.join()
```

```python
'''Q5. State advantages and disadvantages of multithreading.'''

#Ans..

'''Advantages of multithreading:
1. Improved performance: Multithreading allows for concurrent execution of multiple tasks, which can lead 
to improved performance and faster completion of tasks.

2. Resource utilization: Multithreading allows for better utilization of system resources, such as CPU 
and memory, by efficiently distributing tasks among multiple threads.

3. Responsiveness: Multithreading can enhance the responsiveness of an application by allowing it to
continue executing other tasks while waiting for certain operations to complete.

4. Modularity: Multithreading allows for modular design and implementation of complex systems by dividing
them into smaller, more manageable threads that can run concurrently.

5. Enhanced user experience: Multithreading can improve the user experience by providing smoother and more
responsive interfaces, especially in applications that involve heavy computations or I/O operations.

Disadvantages of multithreading:
1. Complexity: Multithreading introduces complexity in terms of synchronization, resource sharing, and 
potential race conditions.

2. Debugging and testing: Debugging and testing multithreaded applications can be more challenging due to
the non-deterministic nature of thread execution and potential concurrency issues.

3. Overhead: Multithreading can introduce overhead in terms of memory usage and context switching between
threads, which can impact performance if not managed efficiently.

4. Scalability limitations: In some cases, multithreading may not provide significant performance benefits
due to limitations in hardware resources or the nature of the task itself.

5. Portability: Multithreading can introduce portability issues, as different operating systems and 
platforms may have different threading models and APIs.'''
```

```python
'''Q6. Explain deadlocks and race conditions.'''

#Ans..

'''Deadlocks:
A deadlock occurs when two or more threads are waiting indefinitely for each other to release resources
that they need to proceed. In other words, it is a situation where two or more threads are stuck in a
circular dependency, preventing any of them from progressing.

Race conditions:
A race condition occurs when multiple threads access and modify shared data concurrently, leading to 
unpredictable and undesirable behavior. Race conditions can happen when the outcome of the program depends
on the relative timing of events between threads.'''
```
