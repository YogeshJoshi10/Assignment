```python
'''Q1. What is an Exception in python? Write the difference between Exceptions and Syntax errors.'''

#Ans..
'''An Exception in Python is an event that occurs during the execution of a program, which disrupts the normal
flow of the program's instructions. When an exception occurs, the program stops executing further 
instructions in the try block and jumps to the appropriate exception handler.

1. Syntax errors: These errors occur when the Python interpreter encounters a statement that violates the 
rules of the Python language. It usually happens due to incorrect syntax, such as missing parentheses,
misspelled keywords, or incorrect indentation.

2. Exceptions: These errors occur during the execution of a program after it has successfully passed the 
parsing phase. Exceptions can happen due to various reasons, such as invalid user input, division by zero,
or accessing a non-existent file. '''
```

```python
'''Q3. Which python statements are used to catch and handle exceptions? Explain with an example.'''

#Ans..

'''The Python statements used to catch and handle exceptions are try and except. 

The try block is used to enclose the code that may raise an exception. If an exception occurs within the
try block, it is caught and the code within the corresponding except block is executed.

For example:'''

a = 10
b = 0

try:
    result = a / b
    print("Result:", result)
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")
```

```python
'''Q4. Explain with an example:
a. try and else
b. finally
c. else'''

#Ans...

'''The else block is used in conjunction with the try block to specify a block of code that should be 
executed if no exceptions are raised within the try block.

for example:'''

x = 10
y = 5

try:
    result = x / y
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")
else:
    print("Result:", result)

    
'''The finally block is used to specify a block of code that should be executed regardless of whether an
exception was raised or not.

for example:'''

python
a = 10
b = 0

try:
    result = a / b
    print("Result:", result)
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")
finally:
    print("This code will always be executed.")
```

```python
'''Q5. What are Custom Exceptions in python? Why do we need Custom Exceptions? Explain with an example.'''

#Ans..

'''Custom exceptions in Python are user-defined exceptions that allow developers to create their own specific 
exception types. They can be used to handle specific error conditions that are not covered by the built-in
exceptions in Python.

There are a few reasons why we may need custom exceptions:

1. Specific Error Handling: Custom exceptions allow us to handle specific error conditions in a more 
meaningful way.

2. Code Organization: By creating custom exceptions, we can organize our code and make it more readable. 
Instead of using generic exceptions like ValueError or TypeError, we can create custom exceptions with
meaningful names that reflect the specific error condition being handled.

3. Code Reusability: Custom exceptions can be reused across multiple projects or modules. Once defined, 
they can be imported and used wherever needed, providing consistency in error handling.

For example:'''

class InsufficientFundsError(Exception):
    def __init__(self, amount, balance):
        self.amount = amount
        self.balance = balance

    def __str__(self):
        return f"Insufficient funds. Available balance: {self.balance}, requested amount: {self.amount}"

def withdraw(amount, balance):
    if amount > balance:
        raise InsufficientFundsError(amount, balance)
    else:
        
        pass

try:
    withdraw(1000, 500)
except InsufficientFundsError as e:
    print(e)
```

```python
'''Q6. Create a custom exception class. Use this class to handle an exception.'''

#Ans..

class CustomException(Exception):
    def __init__(self, message):
        self.message = message

    def __str__(self):
        return f"Custom Exception: {self.message}"

def divide_numbers(a, b):
    if b == 0:
        raise CustomException("Cannot divide by zero")
    else:
        return a / b

try:
    result = divide_numbers(10, 0)
except CustomException as e:
    print(e)
```
